```python
df = pd.read_csv(csv_path)
fnames = list(df['fname'])
fname_ds = tf.data.Dataset.from_tensor_slices(fnames)
waveform_ds = fname_ds.map(get_waveform_and_label)
dset = tf.data.Dataset.zip((fname_ds, waveform_ds))
dset = dset.shuffle(buffer_size=200)
```

```python
rows = 3
cols = 3
n = rows*cols
fig, axes = plt.subplots(rows, cols, figsize=(10, 12))
for i, (fname, (audio, label)) in enumerate(dset.take(n)):
    r = i // cols
    c = i % cols
    ax = axes[r][c]
    ax.plot(audio.numpy())
    ax.set_yticks(np.arange(-1.2, 1.2, 0.2))
    label = label.numpy().decode('utf-8')
    ax.set_title(label)

plt.show()
```

```python
def plot_spectrogram(spectrogram, ax):
    # Convert to frequencies to log scale and transpose so that the time is
    # represented in the x-axis (columns).
    log_spec = np.log(spectrogram.T)
    height = log_spec.shape[0]
    X = np.arange(16000, step=height + 1)
    Y = range(height)
    ax.pcolormesh(X, Y, log_spec)

def get_spectrogram_tf(waveform):
    zero_padding = tf.zeros([16000] - tf.shape(waveform), dtype=tf.float32)
    waveform = tf.cast(waveform, tf.float32)
    equal_length = tf.concat([waveform, zero_padding], 0)
    spectrogram = tf.signal.stft(equal_length, frame_length=255, frame_step=128, fft_length=256)
    spectrogram = tf.abs(spectrogram)
    return spectrogram


def get_spectrogram_librosa(waveform):    
    spectrogram_cfg = {}
    spectrogram_cfg['n_fft'] = 512
    spectrogram_cfg['hop_length'] = 128
    spectrogram_cfg['win_length'] = 512
    spectrogram_cfg['center'] = True
    data_format = type(waveform)
    waveform = waveform.numpy()
        
    spectrogram = librosa.feature.melspectrogram(waveform, **spectrogram_cfg)
    spectrogram = tf.convert_to_tensor(spectrogram, dtype=tf.float32)
    return spectrogram

for fname, (waveform, label) in dset.take(1):
    label = label.numpy().decode('utf-8')
    spectrogram_lib = get_spectrogram_librosa(waveform)
    spectrogram_tf = get_spectrogram_tf(waveform)

plt.figure(figsize=(10, 4))
spectrogram_db = librosa.power_to_db(spectrogram_lib.numpy(), ref=np.max)
librosa.display.specshow(spectrogram_db, x_axis='time', y_axis='mel', sr=16000, fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-frequency spectrogram')
plt.tight_layout()

fig, axes = plt.subplots(2, figsize=(12, 8))
timescale = np.arange(waveform.shape[0])
axes[0].plot(timescale, waveform.numpy())
axes[0].set_title('Waveform')
axes[0].set_xlim([0, 16000])
plot_spectrogram(spectrogram_tf.numpy(), axes[1])
axes[1].set_title('Spectrogram')


plt.show()
```
